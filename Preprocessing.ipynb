```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
iris= pd.read_csv('iris.csv', names=['sepal length', 'sepal width', 'petal length', 'petal width','Label'])
```

```python
iris
```

```python
x= iris.iloc[:,0:4].values
```

```python
x
```

```python
y= iris.iloc[:,4].values
```

```python
y
```

```python
# for encoding dependent variable there are two ways
# 1-label encoder
# 2-OneHot encoder
```

```python
# 1-label encoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
```

```python
y_le= le.fit_transform(y)
```

```python
y_le
```

```python
# 2-OneHot encoder
```

```python
y.shape
```

```python
y= y.reshape(-1,1)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe= OneHotEncoder(sparse=False) 
# sparse-the number of zero elements is much higher than the number of non-zero elements
# False as we dont have zero element
```

```python
y_ohe= ohe.fit_transform(y)
```

```python
y_ohe
```
