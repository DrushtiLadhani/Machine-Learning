```python
data_with_dummies= pd.get_dummies(data_with_no_multi,drop_first=True)
# .get dummies- convert categorical variable into dummy/indicator variables
```

```python
data_with_dummies
```

```python
x= data_with_dummies.drop(['log_price'],axis=1)
```

```python
x
```

```python
y= data_with_dummies['log_price']
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
# Standardscalar is used to standardize the data
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled= scaler.transform(x)
# .transform is used fit and transform on the input data at a single time and converts the data points
```

```python
x_scaled
```

```python
x_scaled.shape
```

```python
from sklearn.model_selection import train_test_split
# train_test_split method is used to split our data into train and test setsx_
```

```python
x_train,x_test,y_train,y_test= train_test_split(x_scaled,y,test_size=0.2,random_state=0)
# tes_size indicates the percentage of the data that should be held over for testing
# so here 20% of data is used for testing and 80% for training
# random_state controls the shuffling process.random_state=0- we get the same train and test sets across different executions
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)    #R2
```

```python
model.intercept_  #b0
```

```python
model.coef_    #b1,b2.......
```

```python
# Testing model
```

```python
y_hat= model.predict(x_test)
```

```python
y_hat
```
