```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
level= np.array([1,2,3,4,5,6])
```

```python
level
```

```python
salary= np.array([100,50,250,70,10,150])
```

```python
salary
```

```python
# (plt.bar=for bar graph, plt.scatter=dot graph, plt.plot=line graph, plt.pie=for pie chart)
```

```python
plt.bar(level,salary,color='black')
plt.xlabel('Level',fontsize=20,color='red')
plt.ylabel('Salary',fontsize=20,color='red')
plt.title('Testing',fontsize=20,color='red')
plt.show()
```

```python
plt.scatter(level,salary,color='black')
plt.xlabel('Level',fontsize=20,color='red')
plt.ylabel('Salary',fontsize=20,color='red')
plt.title('Testing',fontsize=20,color='red')
plt.show()
```

```python
plt.plot(level,salary,color='black')
plt.xlabel('Level',fontsize=20,color='red')
plt.ylabel('Salary',fontsize=20,color='red')
plt.title('Testing',fontsize=20,color='red')
plt.show()
```

```python
plt.plot(level,salary,color='black',marker='*')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.plot(level,salary,color='black',marker='*',markersize=10)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.plot(level,salary,color='black',marker='*',markersize=10,ls='--')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.plot(level,salary,color='black',marker='*',markersize=10,ls='--',markerfacecolor='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.plot(level,salary,color='black',marker='*',markersize=10,ls='--',markerfacecolor='red',lw=5)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Testing')
plt.show()
```

```python
plt.pie(salary)   #( in pie chart only one list can be used )
plt.show()
```
